```python
def make_norm_plot(mu0, color='black', linestyle='-'):
    mu1=mu0+0.5
    mu2=mu0-0.5
    d1 = np.random.normal(loc=mu1,   scale=1, size=1000)
    d2 = np.random.normal(loc=mu2, scale=1, size=1000)
    t1 = np.concatenate((d1,d2))
    sns.distplot(t1, hist=False, color=color,kde_kws={"linestyle":linestyle})
    plt.axis('off')
    for n in range(3):
        plt.plot(mu1+np.random.normal(loc=0, scale=0.2), 0, 'o', markerfacecolor='none', color='black')
        plt.plot(mu2+np.random.normal(loc=0, scale=0.2), 0, 'o', markerfacecolor='none', color='black')
```
